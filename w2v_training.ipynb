```python
import pandas as pd
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
df = pd.read_csv('training_set/pos_tagged.csv')
```

```python
df.head()
```

```python
df['sensed_kalimat'] = df.apply(lambda r: r['kalimat'].replace(r['kata'], f'{r["kata"]}|{r["sense"]}'), axis=1)
```

```python
sw_remover = StopWordRemoverFactory().create_stop_word_remover()
```

```python
uniq_katas = df['kata'].unique()
separated = {kata: df[df['kata'] == kata] for kata in uniq_katas}
```

```python
target_df = separated['cerah']

cerah_sw_removed = target_df['sensed_kalimat'].apply(lambda s: sw_remover.remove(s))
cerah_tokenized = [word_tokenize(s) for s in cerah_sw_removed]
```

```python
cerah_tokenized[160][7] = 'cerah'
```

```python
cerah_tokenized
```

```python
cerah_w2v = Word2Vec(cerah_tokenized, min_count=1, window=3)
```

```python
cerah_w2v.wv.most_similar(positive=['cerah|4801'])
```

```python
cerah_w2v.similarity('cerah', 'cerah|4801')
```

```python
cerah_w2v.similarity('cerah', 'cerah|4802')
```
