```python
import subprocess
import os
import base64
import pandas as pd
```

```python
dataset = pd.read_csv('training_set/cleaned_agreed.csv')
```

```python
os.chdir('/home/aselhidayat/Documents/Academics_6/NLP/indo_tagger/tagger')
```

```python
tagset = {}

for index, row in dataset.iterrows():
    kalimat = row['kalimat']
    kata = row['kata']
    tag_sets = subprocess.check_output(f'./tag.sh -raw "{kalimat}"', shell=True).decode('utf-8').split('\n')
    
    for ts in tag_sets:
        if kata in ts:
            _, tag = ts.split('\t')
            print(kata, tag)
            tagset[index] = tag
            break
    
```

```python
dataset['pos'] = dataset.index.map(tagset)
```
